```python
import os
import csv
from DECIMER import predict_SMILES

class DECIMERProcessor:
    
    
    def __init__(self, image_folder, output_csv):
        self.image_folder = image_folder
        self.output_csv = output_csv
        print(f"Processor ready!")
        print(f"Image Folder: {self.image_folder}")
        print(f"Output CSV:   {self.output_csv}")

    
    def get_smiles(self, image_path):
        try:
            smiles = predict_SMILES(image_path)
            return smiles
        except Exception as e:
            print(f" Error on {image_path}: {e}")
            return "ERROR"

    
    def process_folder(self):
        results = []
        
        for filename in os.listdir(self.image_folder):
            if filename.endswith(".png") or filename.endswith(".jpg"):
                
                full_path = os.path.join(self.image_folder, filename)
                print(f" Processing: {filename}")
                
                smiles = self.get_smiles(full_path)
                results.append([full_path, smiles])
                print(f" Done: {smiles}")
        
        return results

    
    def save_to_csv(self):
        results = self.process_folder()
        
        with open(self.output_csv, mode='w', newline='') as csv_file:
            writer = csv.writer(csv_file)
            writer.writerow(["image_path", "smiles"])
            writer.writerows(results)
        
        print(f"\n CSV saved: {self.output_csv}")
        print(f"Total images processed: {len(results)}")
```

```python
# ================================
# STEP 1 — Import all libraries
# ================================
import os
import csv
import cv2
import numpy as np
from PIL import Image, ImageFilter, ImageEnhance
from DECIMER import predict_SMILES

# ================================
# STEP 2 — Build The Class
# ================================
class DECIMERProcessor:
    
    # ------------------------------
    # BUTTON 0 — Setup The Machine
    # ------------------------------
    def __init__(self, image_folder, output_csv):
        self.image_folder = image_folder
        self.output_csv = output_csv
        print(f" Machine is ready!")
        print(f" Image Folder: {self.image_folder}")
        print(f" Output CSV:   {self.output_csv}")

    # ------------------------------
    # BUTTON 1 — Clean The Image
    # ------------------------------
    def clean_image(self, image_path):
        print(f" Cleaning: {image_path}")
        
        # Open image using OpenCV
        img = cv2.imread(image_path)
        
        # Step 1 — Remove noise and scratches
        denoised = cv2.fastNlMeansDenoisingColored(img, None, 10, 10, 7, 21)
        
        # Step 2 — Convert to grayscale
        gray = cv2.cvtColor(denoised, cv2.COLOR_BGR2GRAY)
        
        # Step 3 — Sharpen the image
        kernel = np.array([[0, -1, 0],
                           [-1, 5,-1],
                           [0, -1, 0]])
        sharpened = cv2.filter2D(gray, -1, kernel)
        
        # Step 4 — Fix brightness and contrast
        clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
        enhanced = clahe.apply(sharpened)
        
        # Step 5 — Save clean image
        clean_path = image_path.replace(".png", "_clean.png")
        cv2.imwrite(clean_path, enhanced)
        
        print(f" Cleaned image saved: {clean_path}")
        return clean_path

    # ------------------------------
    # BUTTON 2 — Get SMILES
    # ------------------------------
    def get_smiles(self, image_path):
        print(f" Reading formula: {image_path}")
        try:
            smiles = predict_SMILES(image_path)
            print(f" SMILES found: {smiles}")
            return smiles
        except Exception as e:
            print(f"Error: {e}")
            return "ERROR"

    # ------------------------------
    # BUTTON 3 — Process All Images
    # ------------------------------
    def process_folder(self):
        print(f"\n Opening folder: {self.image_folder}")
        results = []
        
        # Get all files in folder
        all_files = os.listdir(self.image_folder)
        
        # Filter only images
        images = [f for f in all_files if f.endswith(".png") 
                  or f.endswith(".jpg")]
        
        print(f" Total images found: {len(images)}")
        print("-" * 40)
        
        # Loop through each image
        for i, filename in enumerate(images):
            print(f"\n[{i+1}/{len(images)}] Processing: {filename}")
            
            # Build full path
            full_path = os.path.join(self.image_folder, filename)
            
            # Skip already cleaned images
            if "_clean" in filename:
                print("⏭Skipping clean image...")
                continue
            
            # Button 1 — Clean image first
            clean_path = self.clean_image(full_path)
            
            # Button 2 — Get SMILES from clean image
            smiles = self.get_smiles(clean_path)
            
            # Save result
            results.append([full_path, smiles])
            print("-" * 40)
        
        return results

    # ------------------------------
    # BUTTON 4 — Save To CSV
    # ------------------------------
    def save_to_csv(self):
        print("\n Starting full process...\n")
        
        # Button 3 — Process all images
        results = self.process_folder()
        
        # Save to CSV
        with open(self.output_csv, mode='w', newline='') as csv_file:
            writer = csv.writer(csv_file)
            
            # Header row
            writer.writerow(["image_path", "smiles"])
            
            # All results
            writer.writerows(results)
        
        # Final message
        print(f"\nALL DONE!")
        print(f" Total images processed: {len(results)}")
        print(f"CSV saved at: {self.output_csv}")
```

```python
# Create the machine
processor = DECIMERProcessor(
    image_folder = r"C:\DECIMER.ai\Decimer.ai\image",
    output_csv   = r"C:\DECIMER.ai\Decimer.ai\results.csv"
)

# Press one button — everything is done
processor.save_to_csv()
```
